# Causal transformer: iGPT

One challenge with PixelCNN and its variants is that masked convolutional filters have limited receptive fields, making it difficult to capture long-range dependencies in images. An alternative approach is to use attention-based models. For instance, the Image GPT (iGPT), introduced by OpenAI in the paper "Generative Pretraining from Pixels," uses a transformer architecture to model the conditionals. <br>

To ensure that the autoregressive property is maintained, iGPT uses masked self-attention layers. Before applying the softmax activation to the attention scores, the upper triangular portion of the score matrix is set to a large negative value. This masking ensures that the attention weights are strictly lower triangular, thereby preserving the autoregressive property.

$$
\begin{bmatrix}
s_{1,1} & -\infty & -\infty & \cdots & -\infty \\
s_{2,1} & s_{2,2} & -\infty & \cdots & -\infty \\
s_{3,1} & s_{3,2} & s_{3,3} & \cdots & -\infty \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
s_{D,1} & s_{D,2} & s_{D,3} & \cdots & s_{D,D} \\
\end{bmatrix}
\xrightarrow{\text{Softmax}}
\begin{bmatrix}
0 & 0 & 0 & \cdots & 0\\
\alpha_{2, 1} & 0 & 0 & \cdots & 0\\
\alpha_{3, 1} & \alpha_{3, 2} & 0 & \cdots & 0\\
\vdots & \vdots & \vdots & \ddots & \vdots\\
\alpha_{D, 1} & \alpha_{D, 2} & \alpha_{D, 3} & \cdots & 0\\
\end{bmatrix}
$$

## iGPT Architecture

iGPT uses a decoder only transformer, following the formulation of GPT-2. Each transformer block act on input sequence $\mathbf{h}_l$ as follows

$$
\begin{align*}
    &\mathbf{n}_l = \text{Layer norm}(\mathbf{h}_l)\\
    &\mathbf{a}_l = \mathbf{h}_l + \text{Attention}(\mathbf{h}_l)\\
    &\mathbf{h}_{l+1} = \mathbf{a}_l + \text{MLP}(\text{Layer norm}(\mathbf{a}_l))\\
\end{align*}
$$

For the MLP layers, we used linear layers coupled with GeLU activation functions. In the following section, we will implement iGPT and evaluate it on binarized MNIST data and CIFAR-10 data.

## iGPT Implementation

````python
import torch
import torch.nn as nn
import torch.nn.functional as F
import math
from torchvision import datasets, transforms
import torchvision
from torch.utils.data import DataLoader
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans

DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

class AddTokensTransform:
    def __call__(self, x):
        x = x.view(-1)
        x = (x > 0.5).float()
        x = torch.cat([torch.tensor([2.0]), x])
        return x

batch_size = 32
mnist_transform = transforms.Compose([
    transforms.ToTensor(),
    AddTokensTransform()
])

mnist_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=mnist_transform)
mnist_dataloader = DataLoader(mnist_dataset, batch_size=batch_size, shuffle=True)

cifar_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

cifar_trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=cifar_transform)
cifar_trainloader = torch.utils.data.DataLoader(cifar_trainset, batch_size=64, shuffle=True)
````

### Masked attention implementation

````python
class MultiHeadSelfAttention(nn.Module):
    def __init__(self, embed_dim, n_heads):
        super(MultiHeadSelfAttention, self).__init__()
        self.n_heads = n_heads
        self.embed_dim = embed_dim
        self.W_q = nn.Linear(embed_dim, embed_dim)
        self.W_k = nn.Linear(embed_dim, embed_dim)
        self.W_v = nn.Linear(embed_dim, embed_dim)

    def forward(self, x):
        ## x: [B, L, E], mask: [L, L]
        q = self.W_q(x)
        k = self.W_k(x)
        v = self.W_v(x)
        q = q.reshape(q.shape[0], q.shape[1], self.n_heads, self.embed_dim // self.n_heads)   
        k = k.reshape(k.shape[0], k.shape[1], self.n_heads, self.embed_dim // self.n_heads)  
        v = v.reshape(v.shape[0], v.shape[1], self.n_heads, self.embed_dim // self.n_heads) 
        q = q.transpose(1, 2)
        k = k.transpose(1, 2)
        v = v.transpose(1, 2)
        scores = (q @ k.transpose(-2, -1)) * (1.0 / math.sqrt(k.size(-1)))  
        mask = torch.tril(torch.ones(x.shape[1], x.shape[1], device=x.device, requires_grad=False))
        mask = mask.unsqueeze(0).unsqueeze(0) 
        scores = scores.masked_fill(mask == 0, float('-inf'))
        weights = F.softmax(scores, dim=-1)
        out = torch.matmul(weights, v)   
        out = out.permute(0, 2, 1, 3)    
        out = out.reshape(out.shape[0], out.shape[1], -1)  
        return out
````

### Transformer Block Implementation

````python
class TransformerBlock(nn.Module):

    def __init__(self, embed_dim, n_heads):
        super(TransformerBlock, self).__init__()
        self.pre_layernorm = nn.LayerNorm(embed_dim)
        self.attention = MultiHeadSelfAttention(embed_dim, n_heads)
        self.post_layernorm = nn.LayerNorm(embed_dim)

        self.linear = nn.Sequential(
            nn.Linear(embed_dim, 4 * embed_dim),
            nn.Linear(4 * embed_dim, embed_dim),
            nn.GELU(),
            nn.Dropout(0.1)
        )

    def forward(self, x):
        x = x + self.attention(self.pre_layernorm(x))
        x = x + self.linear(self.post_layernorm(x))
        return x
````

### iGPT Implementation

````python
class Transformer(nn.Module):

    def __init__(self, embed_dim, n_heads, n_layers, n_vocab, max_len, mask=True, device="cpu"):
        super(Transformer, self).__init__()
        self.embedding = nn.Embedding(n_vocab, embed_dim)
        self.postional_embedding = nn.Embedding(max_len, embed_dim)
        self.net = nn.ModuleList(TransformerBlock(embed_dim, n_heads) for _ in range(n_layers))
        self.linear = nn.Linear(embed_dim, n_vocab)
        self.max_len = max_len
        self.device = device

    def forward(self, x):
        ## x: [B, L]
        x = self.embedding(x)  ## [B, L, E]
        pos = torch.arange(x.shape[1], device=self.device).unsqueeze(0)
        x += self.postional_embedding(pos)
        for layer in self.net:
            x = layer(x)  ## [B, L, E]
        x = self.linear(x)  ## [B, L, V]
        return x

    def sample(self, n, dim):
        self.eval()
        with torch.no_grad():
            sample = torch.fill(torch.zeros(n, 1), 2).long().to(DEVICE)
            for i in range(dim**2):
                prob = self.forward(sample)
                prob = F.softmax(prob, dim=-1)
                output = torch.multinomial(prob[:, i], 1)
                sample = torch.cat((sample, output), dim=1)
            samples = sample[:, 1:].view(n, dim, dim)
            samples = samples.detach().cpu().numpy()
        return samples
````

## CIFAR10

We now test iGPT on CIFAR-10 images. To do this, we follow the authors' approach of reducing the context length. The image size used in the paper is \(512 \times 512 \times 3\). Training a transformer directly on such large images would result in logits with dimensions much larger than those in natural language models, making it impractical for GPUs. To address this issue, the authors downsampled the images to \(64 \times 64 \times 3\). Additionally, to further reduce the context length, they applied \(k\)-Means clustering to all pixels in the training set. Each RGB vector was replaced with its nearest cluster center, creating a color palette with 512 clusters, effectively reducing the color representation to a 9-bit palette. We followed this approach for the CIFAR10 data. The processed image is shown below. 

````python
def color_clustering(train_loader):
    mbkmeans = MiniBatchKMeans(n_clusters=512, random_state=42, batch_size=64)
    for X, _ in train_loader:
        X = X.permute(0, 2, 3, 1).reshape(-1, 3).numpy()
        mbkmeans.partial_fit(X)
    return mbkmeans

def get_cluster_index(batch, kmeans):
    batch_size, C, H, W = batch.shape
    reshaped_batch = batch.permute(0, 2, 3, 1).reshape(-1, 3).numpy()
    labels = kmeans.predict(reshaped_batch)
    labels = labels.reshape(batch_size, H*W)
    return labels, batch_size, C, H, W

def get_image_from_index(labels, kmeans, batch_size, C, H, W):
    labels = labels.reshape(-1)
    images = kmeans.cluster_centers_[labels].reshape(batch_size, H, W, C)
    images = torch.tensor(images).permute(0, 3, 1, 2)
    return images
````

````python
mbkmeans = color_clustering(trainloader)
````

````python
X, _ = next(iter(trainloader))
labels, batch_size, C, H, W = get_cluster_index(X, mbkmeans)
X = get_image_from_index(labels, mbkmeans, batch_size, C, H, W).permute(0, 2, 3, 1)
fig, ax = plt.subplots(6, 6, figsize=(6, 6))
for i in range(6):
    for j in range(6):
        ax[i][j].imshow((X[6*i+j] + 1)/ 2)
        ax[i][j].axis('off')
plt.subplots_adjust(wspace=0.1, hspace=0.1)
plt.show()
````

````python
def preprocess(X, kmeans):
    labels = get_cluster_index(X, kmeans)[0]
    labels = torch.cat((torch.tensor([[512]]).repeat(labels.shape[0], 1), torch.tensor(labels)), dim=1)
    return labels

def plot_samples(samples, m=1, n=6, title="Generated samples"):
    fig, ax = plt.subplots(m, n, figsize=(10, 2))
    ax = ax.flatten()
    for i in range(m*n):
        ax[i].imshow(samples[i], cmap="gray")
        ax[i].axis("off")
    fig.suptitle(title, fontsize=12)
    plt.tight_layout()
    plt.show()

def train_epoch(model, dataloader, optimizer, criterion, kmeans, n_classes, sample=True, epoch_num=0):
    train_loss = 0.0
    for X, _ in dataloader:
        optimizer.zero_grad()
        X = preprocess(X, kmeans)
        X = X.to(DEVICE).long()
        output = model(X[:, :-1])
        loss = criterion(output.reshape(-1, n_classes), X[:, 1:].reshape(-1))
        train_loss += loss.item()
        loss.backward()
        optimizer.step()
    train_loss /= len(dataloader)
    if sample:
        with torch.no_grad():
            sample = torch.full((6,), 512).unsqueeze(1).long().to(DEVICE)
            for i in range(32**2):
                prob = model(sample)
                prob = F.softmax(prob, dim=-1)
                output = torch.multinomial(prob[:, i], 1)
                sample = torch.cat((sample, output), dim=1)
            samples = sample[:, 1:].view(6, -1)
            samples = samples.detach().cpu().numpy()
            samples[samples == 512] = 0
            samples = get_image_from_index(samples, mbkmeans, 6, 3, 32, 32).permute(0, 2, 3, 1)
            samples = (samples + 1) / 2
            plot_samples(samples, title=f"Epoch {epoch_num} generated samples with loss: {train_loss}")
    return train_loss

def train(model, dataloader, optimizer, criterion, kmeans, n_classes, epochs=10):
    for epoch in range(epochs):
        train_epoch(model, dataloader, optimizer, criterion, kmeans, n_classes, epoch_num=epoch+1)
````

For the CIFAR-10 data, we train an iGPT model with an embedding dimension of $128$, $4$ attention heads, and $8$ layers of transformer blocks. The model is trained for 30 epochs using consistent hyperparameters. The generated image is shown below. We observe that the model produces images that resemble CIFAR-10 samples, although some generated images may lack clarity.

````python
model = Transformer(embed_dim=128, n_heads=4, n_layers=8, n_vocab=513, max_len=32*2, device=DEVICE).to(DEVICE)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
train(model, trainloader, optimizer, criterion, mbkmeans, n_vocab, epochs=10)
````
